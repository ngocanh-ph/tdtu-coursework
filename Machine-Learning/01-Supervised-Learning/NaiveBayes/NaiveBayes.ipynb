```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB, ComplementNB, CategoricalNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, Binarizer, KBinsDiscretizer
column_names = [
    'id',
    'diagnosis',
    'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
    'smoothness_mean', 'compactness_mean', 'concavity_mean',
    'concave_points_mean', 'symmetry_mean', 'fractal_dimension_mean',
    'radius_se', 'texture_se', 'perimeter_se', 'area_se',
    'smoothness_se', 'compactness_se', 'concavity_se',
    'concave_points_se', 'symmetry_se', 'fractal_dimension_se',
    'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst',
    'smoothness_worst', 'compactness_worst', 'concavity_worst',
    'concave_points_worst', 'symmetry_worst', 'fractal_dimension_worst'
]
wdbc_df = pd.read_csv('wdbc.data', header=None, names=column_names)
X = wdbc_df.drop(['id', 'diagnosis'], axis=1)
y = wdbc_df['diagnosis']
# Encode label
le = LabelEncoder()
y_encoded = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded,test_size=0.2, random_state=42, stratify=y_encoded)
results = []

# Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
results.append(('Gaussian Naive Bayes', accuracy_score(y_test, y_pred)))

# Chuẩn hóa cho Multinomial và Complement
scaler = MinMaxScaler()
X_train_m = scaler.fit_transform(X_train)
X_test_m = scaler.transform(X_test)

# Multinomial Naive Bayes
mnb = MultinomialNB()
mnb.fit(X_train_m, y_train)
y_pred = mnb.predict(X_test_m)
results.append(('Multinomial Naive Bayes', accuracy_score(y_test, y_pred)))

# Bernoulli Naive Bayes
binarizer = Binarizer(threshold=0.5)
X_train_b = binarizer.fit_transform(X_train_m)
X_test_b = binarizer.transform(X_test_m)
bnb = BernoulliNB()
bnb.fit(X_train_b, y_train)
y_pred = bnb.predict(X_test_b)
results.append(('Bernoulli Naive Bayes', accuracy_score(y_test, y_pred)))

# Complement Naive Bayes
cnb = ComplementNB()
cnb.fit(X_train_m, y_train)
y_pred = cnb.predict(X_test_m)
results.append(('Complement Naive Bayes', accuracy_score(y_test, y_pred)))

#CategoricalNB
kbd = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
X_train_cate = kbd.fit_transform(X_train)
X_test_cate = kbd.transform(X_test)
catnb = CategoricalNB()
catnb.fit(X_train_cate, y_train)
y_pred = catnb.predict(X_test_cate)
results.append(('Categorical Naive Bayes', accuracy_score(y_test, y_pred)))

# So sánh kết quả
compare_df = pd.DataFrame(results, columns=['Model', 'Accuracy'])
print(compare_df)
```

Nhận xét:
- Gaussian Naive Bayes: Đạt độ chính xác cao nhất vì giả định phân phối chuẩn phù hợp với dữ liệu liên tục của WDBC.
- Multinomial Naive Bayes: Độ chính xác thấp hơn do mô hình này phù hợp với dữ liệu dạng đếm không phù hợp với dữ liệu liên tục.
- Bernoulli Naive Bayes: Kết quả khá tốt nhưng bị giảm hiệu năng do dữ liệu phải nhị phân hóa làm mất thông tin.
- Complement Naive Bayes: Cho kết quả thấp vì được thiết kế cho dữ liệu mất cân bằng và dạng đếm không phù hợp với WDBC.
- Categorical Naive Bayes: Đạt độ chính xác cao nhờ rời rạc hóa dữ liệu nhưng vẫn kém Gaussian NB do mất thông tin liên tục.
